## Importing modules in Python: Modules and Packages

### - In Python,modules and packages, helps organize and reuse codes.
### - Modules are simply files that contain Python code (functions, variables, classes) which you can reuse in other programs.
### -  How to import them 


```python
# Types of Modules:

# 1. Built-in Modules:
# These come with Python already installed.
# Examples:
# math
# random
# datetime

# Example:
# import math
# print(math.sqrt(16))  # Output: 4.0

# 2. User-defined Modules:
# Modules you create yourself.
# # mymath.py
# def add(a, b):
#     return a + b
# Usage:
# import mymath
# print(mymath.add(2, 3))

# 3. External Modules (Libraries):
# Installed using pip.
# Examples:
# numpy
# pandas
# requests
# pip install numpy
# import numpy as np
# print(np.array([1, 2, 3]))
```

## 🔹 Ways to Import Modules

```python
# 1. Import entire module
# import math
# print(math.pi)

# 2. Import specific items
# from math import sqrt
# print(sqrt(25))

# 3. Import with alias
# import math as m
# print(m.pi)
```

```python
import random
print(random.randint(1, 10))  # Output: Random integer between 1 and 10
print(random.randint(2,30))




# l=[23,45,11,10,15]
l=['pink','orange','blue','green','yellow','grey','magenta','maroon']
print(random.choice(l))
print(random.choices(l,k=3))  #k is number of items to choose
```

```python
# csv

import csv
with open('example.csv', mode='w', newline='') as file:    #to write to a csv file (mode='w' for writing, newline='' to avoid extra blank lines)
    writer = csv.writer(file)
    writer.writerow(['Name', 'Age'])
    writer.writerow(['Alice', 30])
    writer.writerow(['Bob', 25])

with open('example.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)  
```

```python
# Regular Expressions
import re
text = "The rain in Spain stays mainly in the plain."
pattern = r'\b\w+ain\b'  #to find words that end with 'ain'
matches = re.findall(pattern, text)
print(matches)  # Output: ['rain', 'Spain', 'plain']


```
