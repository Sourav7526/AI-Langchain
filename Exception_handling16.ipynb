## Understanding Exceptions
- Exception handling in python allows you to handle errors gracefully and take corrective actions without stopping the execution of the program.

## What are Exceptions?
- Exceptions are events that occur during the execution of a program that disrupt the normal flow of a program.
They occur when an error is encountered, such as trying to divide by zero or accessing an index that is out of range.

### Here we will learn how to use try,except,else and finally to handle exceptions in python.

. ZeroDivisionError: This exception is raised when you try to divide a number by zero.
. FileNotFoundError: This exception is raised when you try to access a file that does not exist.
. ValueError: This exception is raised when you try to convert a string to a number, but the string does not contain a valid number.
. IndexError: This exception is raised when you try to access an index that is out of range in a list or a string.
. TypeError: This exception is raised when you try to perform an operation on a value of the wrong type, such as trying to add a string and an integer.

```python
try:
    num = int(input("Enter a number: "))
    result = 10/num
except ValueError as ve:
    print(ve)
    print("Please enter a valid integer")
except ZeroDivisionError as ze:
    print(ze)
    print("Denominator cannot be zero")
# except Exception as ex:
#     print(ex)
#     print("An unexpected error occurred")
```

```python
#try ,except else block
try:
    num = int(input("Enter a number: "))
    result = 10/num
except ValueError as ve:
    print(ve)
    print("Please enter a valid integer")
except ZeroDivisionError as ze:
    print(ze)
    print("Denominator cannot be zero")
except Exception as ex:
    print(ex)
    print("An unexpected error occurred")
else:
    print(f"The result is: {result}")
    
```

```python
#try ,except, else and finallyblock
try:
    num = int(input("Enter a number: "))
    result = 10/num
except ValueError as ve:
    print(ve)
    print("Please enter a valid integer")
except ZeroDivisionError as ze:
    print(ze)
    print("Denominator cannot be zero")
except Exception as ex:
    print(ex)
    print("An unexpected error occurred")
else:
    print(f"The result is: {result}")
finally:
    print("Execution Complete")
```

- try: Runs the risky code that might cause an error.
- except: Catches and handles the error if one occurs.
- else: Executes only if no exception occurs in try.
- finally: Runs regardless of what happens useful for cleanup tasks like closing files.

```python
#File handling with exception handling
try:
    with open("non_existent_file.txt", "r") as file:
        content = file.read()
        print(content)
except FileNotFoundError as fn:
    print("File does not exist. Please check the file name and path.")
finally:
    if 'file' in locals() and not file.closed():
        file.close()
        print("File closed successfully")
```
