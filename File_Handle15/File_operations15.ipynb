## File Operations - Reading and Writing Files in Python
File handling is a crucial part of any programming language. Python provides built-in functions and methods to read and write files in both text and binary modes. In this section, we will explore how to perform file operations in Python.
In simple terms, file handling allows you to create, read, update, and delete files on your computer using Python. This is essential for tasks such as data storage, logging, and configuration management.

```python
##Syntax
# file = open('filename.txt', 'mode')


# filename.txt: name (or path) of the file to be opened.
# mode: mode in which you want to open the file (read, write, append, etc.).
```

```python
## Read Whole file
with open('Example.txt', 'r') as file:
     #'r' stands for read mode                  
    content = file.read()                     #read the whole file content
    print(content)
```

```python
## Read a file line by line
with open('Example.txt', 'r') as file:
    for line in file:   
        print(line)                      #iterating through each line in the file
        # print(line.strip())                   #strip() removes leading/trailing whitespace
```

```python
# Writing to a file(overwriting)
with open('Example.txt', 'w') as file:
    #'w' stands for write mode (overwrites existing content)
    file.write("This is a new file created using Python.\n")
    file.write("It contains multiple lines of text.\n")
    file.write("File handling is easy in Python.\n")
```

```python
# Writing to a file(without overwriting)
with open('Example.txt', 'a') as file:
    #'a' stands for append mode
    file.write("\nThis line will be added to the existing file.\n")
```

```python
#Writing a list of lines to a file
lines = ["Line 1: This is the first line.\n",
         "Line 2: This is the second line.\n",
         "Line 3: This is the third line.\n"]   
with open('Example.txt', 'a') as file:
    file.writelines(lines)  #writelines() writes a list of lines to the file
```

```python
# Read the content from a source file and write it to a destination file
source_file = 'Example.txt'
destination_file = 'Copy_of_Example.txt'

with open(source_file, 'r') as src:
    content = src.read()  # Read the content of the source file


with open(destination_file, 'w') as dest:
    dest.write(content)  # Write the content to the destination file
```

```python
#writing and then reading from the same file
with open('Example.txt', 'w+') as file:
    file.write("This is a new line added to the file.\n")


    # file.seek(0)  
    # Move the cursor back to the beginning of the file


    content = file.read()  # Read the content of the file
    print(content)
```

```python
# Reading and then writing to the same file
with open('Example.txt', 'r+') as file:
    content = file.read()  # Read the existing content of the file
    print("Existing Content:\n", content)

    file.write("\nThis line is added after reading the file.")  # Write new content to the file
    file.close()
    
```

```python
#Checking File Properties
#Once the file is open, we can check some of its properties:

f = open("Example.txt", "r")
print("Filename:", f.name)
print("Mode:", f.mode)
print("Is Closed?", f.closed)

f.close()
print("Is Closed?", f.closed)
```
